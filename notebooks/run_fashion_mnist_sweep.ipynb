```
# Show results
import pandas as pd, glob

csvs = sorted(glob.glob('results_real/raw/*fashion_mnist*.csv'))
print(f"Completed: {len(csvs)} experiments\n")

df = pd.concat([pd.read_csv(f) for f in csvs], ignore_index=True)
pivot = df.pivot_table(values='snn_accuracy', index='model', columns='T', aggfunc='mean')
print((pivot * 100).round(2).to_string())
print(f"\nANN accuracy: {df['ann_accuracy'].mean()*100:.2f}%")
```
